```python
df1 = pd.DataFrame(df, columns=['street'])

# Define regular expressions to capture different parts of the addresses
patterns = {
    'street_number' : r'^(\d{1,5}(?:-\d{1,5})?)',
    'street_name':r'^\d{1,5}(?:-\d{1,5})?\s+([A-Za-z\s]+?)\s+\d{1,2}(?:st|nd|rd|th)$',
    'street_type': r'\d{1,2}(?:st|nd|rd|th)$',
    'direction': r'\b(N|S|E|W|NE|NW|SE|SW)\b'
}

# Function to extract parts based on the regex
def extract_part(pattern, text):
    match = re.search(pattern, text)
    return match.group(1) if match else None

# Apply the regex patterns to extract parts into new columns
for key, pattern in patterns.items():
    df1[key] = df['street'].apply(lambda x: extract_part(pattern, x))
```

##to handle range values

```python
def preprocess_range(value):
    if value is None:  # Handle None values
        return value
    # Check if the value is a range (improved regex)
    range_match = re.match(r'(\d+)\s*-\s*(\d+)', value)  # Handles potential spaces
    if range_match:
        # Extract the start and end of the range
        start = int(range_match.group(1))
        end = int(range_match.group(2))
        # Calculate the average of the range
        average = (start + end) / 2
        return str(int(average))  # Convert average back to string
    else:
        return value

# Apply the preprocess_range function to the street column
df['street_number'] = df['street_number'].apply(preprocess_range)
```

## check missing values in an array

```python
# Example NumPy array with some direction values and missing values
array = np.array(x[ :,-1])

# Function to check for missing values (either empty strings or NaN strings)
def is_missing(val):
    if val == '' or val == 'nan' or (isinstance(val, float) and np.isnan(val)):
        return True
    return False

# Vectorize the function so it can be applied to each element in the array
vectorized_is_missing = np.vectorize(is_missing)

# Apply the function to the array
missing_values = vectorized_is_missing(array)

# Count the number of missing values
missing_count = np.sum(missing_values)

print("Array:")
print(array)
print("\nMissing values (True indicates a missing value):")
print(missing_values)
print("\nCount of missing values:")
print(missing_count)
```

```python
import numpy as np

# Example NumPy array with some non-numeric strings and None values
array = np.array(x)

# Function to convert values to float or np.nan if conversion fails
def to_float(val):
    try:
        if val is None:
            return np.nan
        return float(val)
    except ValueError:
        return np.nan

# Vectorize the function so it can be applied to each element in the array
vectorized_to_float = np.vectorize(to_float)

# Apply the function to the array
array = vectorized_to_float(array)

# Check for missing values (NaNs) in the array
missing_values = np.isnan(array)

print("Array:")
print(array)
print("\nMissing values (True indicates a NaN):")
print(missing_values)
```
